```python
import pandas as pd

df = pd.read_csv("C:/Users/skabe/OneDrive/Documents/Rudder_Works/SCHOOL/MUT[I.T]/Technical Programming II/2024/TEST 2/bky6-3za5.csv")

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

population = df['ward_id'].values
insertion_sort(population)
df['ward_id'] = population

print("DataFrame after sorting 'Ward ID' in ascending order:")
print(df)
```
