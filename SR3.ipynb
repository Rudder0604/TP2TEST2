```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r'C:\Users\skabe\OneDrive\Documents\Rudder_Works\SCHOOL\MUT[I.T]\Technical Programming II\2024\TEST 2\bky6-3za5.csv'
df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
plt.bar(df['density_dwellings'][:5], df['structure_count'][:5], color='blue')  
plt.xlabel('Structure Count')
plt.ylabel('Density Dwellings')
plt.title('Bar Graph for First Few Rows')
plt.xticks(rotation=45)  
plt.show()
```

```python
#PieChart
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\skabe\OneDrive\Documents\Rudder_Works\SCHOOL\MUT[I.T]\Technical Programming II\2024\TEST 2\bky6-3za5.csv'
df = pd.read_csv(file_path)

category_column = 'structure_count'  
value_column = 'ward_id'       

num_rows = 5
df_first_few_rows = df.head(num_rows)

plt.figure(figsize=(8, 8))
plt.pie(df_first_few_rows[value_column], labels=df_first_few_rows[category_column], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of First Few Rows')
plt.show()


```

```python
#Histogram 
plt.figure(figsize=(10, 6))
sns.histplot(df['structure_count'], kde=True, bins=10)
plt.title('Distribution of Density Dwelling')
plt.xlabel('Population Density Dwelling')
plt.ylabel('Frequency')
plt.show()
```
